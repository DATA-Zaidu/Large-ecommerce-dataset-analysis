```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

df = pd.read_csv(r"large_ecommerce_dataset.csv")

df.head()

## Data cleaning



df.isnull().sum()

df.dropna()

df.drop_duplicates()

df.head()

## Analysis start

# Total_revenue

Total_revenue = df["net_revenue"].sum()

Total_revenue

# Total_orders

Total_orders = df["order_id"].nunique()

Total_orders

# Average order value

Aov = Total_revenue / Total_orders

Aov

# Return orders

return_rate = (df['is_return'].sum() / Total_orders) * 100

return_rate


#cod orders

cod_rate = (df[df['payment_method']==("COD")].shape[0] / Total_orders) * 100

cod_rate


# state wise revenue

state_revenue = df.groupby("state")["net_revenue"].sum().sort_values(ascending=False)

state_revenue

# category wise revenue

category_revenue = df.groupby("product_category")["net_revenue"].sum().sort_values(ascending=False)

category_revenue


# monthly revenue

monthly_revenue = df.groupby("order_month")["net_revenue"].sum().sort_values(ascending=False)

monthly_revenue


# Payment method

payment_method = df.groupby("payment_method")["net_revenue"].value_counts(normalize=True)*100

payment_method

# return rate per category

rrc = df.groupby("product_category")["is_return"].value_counts(normalize=True)*100
rrc
```
